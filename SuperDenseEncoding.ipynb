```python
import qiskit as q
from qiskit import IBMQ
from qiskit.tools.monitor import job_monitor
from qiskit.visualization import plot_bloch_multivector, plot_histogram
```

```python
IBMQ.load_account()
```

```python
provider = IBMQ.get_provider("ibm-q")
```

```python
backend = provider.get_backend("ibmq_qasm_simulator")
sv_backend = q.Aer.get_backend("statevector_simulator")
```

```python
circuit = q.QuantumCircuit(4,4)
circuit.x(0)
circuit.x(1)
circuit.h(2)
circuit.cx(2,3)
circuit.barrier()
circuit.cx(1,2)
circuit.cz(0,2)
circuit.barrier()
circuit.cx(2,3)
circuit.h(2)
circuit.barrier()
circuit.measure([0,1,2,3],[0,1,2,3])
circuit.draw(output="mpl")
```

```python
job = q.execute(circuit, backend=backend, shots=1024)
job_monitor(job)
result = job.result()
```

```python
plot_histogram([result.get_counts(circuit)])
```

```python
sv_job = q.execute(circuit, backend=sv_backend, shots=1024)
statevector = sv_job.result().get_statevector()
plot_bloch_multivector(statevector)
```
